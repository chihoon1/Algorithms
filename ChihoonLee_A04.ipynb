```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.datasets import load_iris
import seaborn as sns
import warnings
import random
warnings.filterwarnings("ignore")
```

```python
data = load_iris()
df = pd.DataFrame(data['data']).iloc[:,:2]
df = df.drop_duplicates()
```

```python
#indices = np.random.randint(len(df), size=25)
#indices = random.sample(range(len(df)), 25)
#indices = np.random.choice(range(len(df)), 25, replace=False)

df = df.iloc[0:25, :2]
df
```

```python
l2_dist_matrix = np.empty((len(df), len(df)))
l2_dist_matrix[:] = np.nan
l2_dist_list = []
#print(l2_dist_metric)
for i in range(len(df)):
    for j in range(i+1, len(df)):
        dist_x_y = np.linalg.norm(df.iloc[i,:]-df.iloc[j,:])
        l2_dist_matrix[i,j] = dist_x_y
        l2_dist_list.append(dist_x_y)

l2_dist_list.sort()
np.set_printoptions(precision=3)
#print(np.array(l2_dist_list))  # uncomment this to see wthe ordered l2 distance between all data points
```

```python
sns.scatterplot(x=df.iloc[:,0], y=df.iloc[:,1])
```

Data set I used is based on first 25 data points from the Iris data set. Also, I only took the first two features from the Iris data set for the simplicity for plotting the data.
By looking at the scatter plot, there seem to be 2 or 3 clusters in the data. For k-means clustering, a potentially good initial means is {(5.1, 3.5),(5.8, 4.0), or if k=3 (4.4, 2.9)}, and potentially bad inital means {(5.4, 3.9), (5.4	3.7), or if k=3 (5.4, 3.4)}. I will explore both sets of the initial means for K-means clustering.
Five relevant clustering distance r for single linkage would be {0.224, 0.316, 0.412, 0.5, 0.64} as those distances appear repeatedly in many edges. They will be the choice of r for single linkage clustering in my codes.

```python
df_k2_1 = df.copy()
k2_mean1 = np.array([[5.1, 3.5], [5.8, 4.0]])
kmeans = KMeans(n_clusters=2, init=k2_mean1).fit(df)
df_k2_1['cluster label'] = kmeans.labels_
sns.scatterplot(x=0, y=1, data=df_k2_1, hue="cluster label")
print(f"Variance is {kmeans.inertia_}")
converged_means = kmeans.cluster_centers_
print(f"Final means:\n{converged_means}")
'''
variance = 0
for i in range(len(df_k2_1)):
    pt, cluster = np.array(df_k2_1.iloc[i,:2]), np.array(df_k2_1.iloc[i,2])
    #print(pt, cluster)
    variance += (pt[0] - converged_means[cluster][0])**2 + (pt[1] - converged_means[cluster][1])**2
print(variance)
'''
del df_k2_1
```

For this one, I used 2 clusters k-means algorithm with initial means (5.1, 3.5) and (5.8, 4.0).
The variance of the final cluster is 3.25114. Centroids after the convergence are {cluster0: (4.879, 3.289), cluster1: (5.517, 3.933)}. Also, cluster1 ended up containing fewer data points than cluster0.

```python
df_k3_1 = df.copy()
k3_mean1 = np.array([[5.1, 3.5], [5.8, 4.0], [4.4, 2.9]])
kmeans = KMeans(n_clusters=3, init=k3_mean1).fit(df)
df_k3_1['cluster label'] = kmeans.labels_
sns.scatterplot(x=0, y=1, data=df_k3_1, hue="cluster label")
print(f"Variance is {kmeans.inertia_}")
converged_means = kmeans.cluster_centers_
print(f"Final means:\n{converged_means}")
del df_k3_1
```

For this one, I used 3 clusters k-means algorithm with initial means (5.1, 3.5),(5.8, 4.0), and (4.4, 2.9).
With k=3, k-means clustering look better. The variance is reduced to 1.7328. Centroids after the convergence are {cluster0: (5.055, 3.509), cluster1: (5.6, 3.96), cluster2: (4.689, 3.078)}. Interestingly, the centroids for cluster0 and 1 are different from k=2 centroids although they started with the same initial means. Also, the number of the data points in each cluster is similar to each other.

```python
df_k2_2 = df.copy()
k2_mean2 = np.array([[5.4, 3.9], [5.4, 3.7]])
kmeans = KMeans(n_clusters=2, init=k2_mean2).fit(df)
df_k2_2['cluster label'] = kmeans.labels_
sns.scatterplot(x=0, y=1, data=df_k2_2, hue="cluster label")
print(f"Variance is {kmeans.inertia_}")
converged_means = kmeans.cluster_centers_
print(f"Final means:\n{converged_means}")
del df_k2_2
```

For this one, I used 2 clusters k-means algorithm with initial means (5.4, 3.9) and (5.4, 3.7).
Unlike my expectation, setting initial means to close to each other didn't have effect on producing worse clustering in terms of variance. Current cluster produced the variance same as the first k-means cluster(k=2). The current cluster converged with the centroids {cluster0: (5.517, 3.933), cluster1: (4.879, 3.289)}.

```python
df_k3_2 = df.copy()
k3_mean2 = np.array([[5.4, 3.9], [5.4, 3.7], [5.4, 3.4]])
kmeans = KMeans(n_clusters=3, init=k3_mean2).fit(df)

df_k3_2['cluster label'] = kmeans.labels_
sns.scatterplot(x=0, y=1, data=df_k3_2, hue="cluster label")
print(f"Variance is {kmeans.inertia_}")
converged_means = kmeans.cluster_centers_
print(f"Final means:\n{converged_means}")
del df_k3_2
```

For this one, I used 3 clusters k-means algorithm with initial means (5.4, 3.9), (5.4, 3.7), and (5.4, 3.4).
Unlike my expectation, the second initial means resulted in better clustering than the first initial means when k=3. Moreover, this k=3 clustering with the second initial means is better than k=2 clustering with the second inital means because the variance is reduced to 1.6789 from 3.2511 in the one above. Centroids after the convergence are {cluster0: (5.65, 4.025), cluster1: (5.16, 3.53), cluster2: (4.691, 3.155)}.  Also, cluster0 has fewer the number of data points in it compared to the the other two clusters.

```python
eps = 1e-7
# added eps to r because python sklearn library defines r < ri for r stopping condition
r = np.array([0.224, 0.316, 0.412, 0.5, 0.64]) + eps
cluster_models = []
for i in range(len(r)):
    cluster_models.append(
        AgglomerativeClustering(distance_threshold=r[i],
                                n_clusters=None).fit(df)
    )
```

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.set_style("whitegrid")
df_r1 = df.copy()
df_r1['cluster label'] = cluster_models[0].labels_
sns.scatterplot(x=0, y=1, data=df_r1,
                hue="cluster label", s=200).set_title(f"r<{r[0]}")
del df_r1
```

```python
df_r2 = df.copy()
df_r2['cluster label'] = cluster_models[1].labels_
sns.scatterplot(x=0, y=1, data=df_r2,
                hue="cluster label", s=200).set_title(f"r<{r[1]}")
del df_r2
```

The choice of r I used for single linkage is {0.224, 0.316, 0.412, 0.5, 0.64}. The output of the clusters produced with the five different cluster distance r verifies the theoretical part of the single linkage clustering because cluster with smaller r refines cluster with bigger r. For instance, for all clusters, a cluster from clustering with r=0.5 is contained in some cluster from clustering with r=0.64. This refinement characteristic adds interpretation in the hierarchical clustering since it tells similarity between blocks in one cluster label.
