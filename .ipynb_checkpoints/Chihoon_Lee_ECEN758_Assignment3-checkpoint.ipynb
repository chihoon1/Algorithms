```python
#problem 1(a)
#inputs of the problem
X1 = np.array([0, 0, 2, 5, 5])
X2 = np.array([2, 0, 0, 0, 2])
D = np.column_stack((X1, X2))

D
```

```python
def initialize_centroid(*args):
    # parameter: initial randomly guessed centroids(must be passed in the ascending order of cluster_i)
    # return a list of numpy array(centroids)
    centroid_list = []  # index of the list corresponds to the cluster identification(Cluster_i)
    for centroid in args:
        centroid_list.append(np.array(centroid))
    return centroid_list
```

```python
def finding_k_means_cluster(D, clusters_list, centroid_list):
    '''
    param: D = data set, clusters_list = initial list of clusters(list of sets)
    , centroid_list = initial list of centroid(list of np.arrys)
    return: tuple containing cluster assignments, centroids, SSE of clusters, and iteration
    '''
    iteration = 0
    epsilon = 0  # preferred sum of squared difference between i-th and i-1-th iteration for convergence
    sum_of_squared = 10**9  # arbitrary number greater than epsilon for first iteration
    while (sum_of_squared > epsilon):  # iteration criteria
        iteration += 1
        # Cluster Assignment for each xj in D
        temp_clusters_list = []
        for i in range(len(centroid_list)):
            temp_clusters_list.append(set())  # can't do []*len() because it copies the same set object 
        temp_centroid_list = [0]*len(centroid_list)
        for j in range(D.shape[0]):
            xj = D[j,:]  # xj is a sample data point(row of D)
            min_distance = 10**9
            for i in range(len(centroid_list)):
                mu_i = centroid_list[i]# mu_i is the centroid of cluster_i
                # the closest cluster by taking min (squared of l-2 norm of (xj minus each centroid))
                dist_xj_mu_i = np.linalg.norm(xj - mu_i)**2
                if dist_xj_mu_i < min_distance:
                    cluster_id = i
                    min_distance = dist_xj_mu_i
            #print("{cluster_id} and dist: {min_distance}".format(cluster_id=cluster_id, min_distance=min_distance))
            temp_clusters_list[cluster_id].add(j)
        # update centroid
        for i in range(len(temp_clusters_list)):
            cluster = temp_clusters_list[i]
            total = np.array([0]*D.shape[1])
            for j in cluster:
                total += D[j,:]
                #print(f"total: {total}")
            temp_centroid_list[i] = total / len(cluster)
            print(f"updated cluster: {cluster}")
        print(f"updated centroid: {temp_centroid_list}")
        # Measure the sum of squared difference between the current iteration and the previous iteration
        sum_of_squared = 0
        for i in range(len(centroid_list)):
            sum_of_squared += np.linalg.norm(temp_centroid_list[i] - centroid_list[i])**2
        centroid_list = temp_centroid_list
        clusters_list = temp_clusters_list
    # computing the SSE of clusters
    SSE = 0
    for i in range(len(centroid_list)):
        for j in clusters_list[i]:
            SSE += np.linalg.norm(D[j,:] - centroid_list[i])**2
    return clusters_list, centroid_list, SSE, iteration
```

```python
#initalization steps
k = 2  # number of clusters
clusters_list = [set()]*k  # index of the list corresponds to the cluster identification(Cluster_i)
centroid_list = initialize_centroid(D[0,:], D[1, :])
print(f"initial cluster list: {clusters_list}")
print(f"centroid list:\n{centroid_list}")
```

```python
clusters_list, centroid_list, SSE, iteration = finding_k_means_cluster(D, clusters_list, centroid_list)
```

```python
from math import pi, sqrt, exp
from scipy.stats import multivariate_normal

def prob_xj_given_theta(xj, mu, sigma):
    '''
    param: xj=a data point(vector), mu=mean(vector), sigma=covariance matrix(2d np array)
    return: Pr(xj | mu, sigma) from normal distribution
    '''
    # print(f"prob debug: {xj}, {mu}, {sigma}")
    # print(f"scipy: {multivariate_normal.pdf(xj, mean=mu, cov=sigma, allow_singular=True)}")
    d = sigma.shape[0]
    centered = (xj - mu)
    inverse_sigma = np.linalg.inv(sigma)

    # I took the absolute value of determinant because covariance matrix is positive semi-definite.
    # So, det always >= 0
    det_sigma = abs(np.linalg.det(sigma))
    const_coeff = sqrt(((2 * pi) ** d) * det_sigma)
    exponent = np.inner(inverse_sigma.T @ centered, centered) # centered@inverse_sigma@centered
    # print(f"exponent: {exponent}")
    try:
        e_exp = exp(-exponent / 2)
        prob = e_exp / const_coeff
    except OverflowError:
        # e^exponent is too large and go over memory allocation for the python double data type
        # In this case, I'll use multivariate_normal.pdf from scipy
        prob = multivariate_normal.pdf(xj, mean=mu, cov=sigma, allow_singular=True)

    if multivariate_normal.pdf(xj, mean=mu, cov=sigma, allow_singular=True) - prob > 0.00001:  # debugging purpose
        print(f"prob diff: {multivariate_normal.pdf(xj, mean=mu, cov=sigma, allow_singular=True) - prob}")
    return prob
```

```python
def expectation_maximization(D, clusters_mean_list, clusters_covmat_list, prior_prob_list, epsilon=0.001, stop=100):
    '''
    param: D=data(2d np array), cluster_mean_list=initial mu for each cluster(index=0:n-1 represent each cluster),
    clusters_covmat_list=initial covariance matrix for each cluster
    prior_prob_list=initial prior probabilities for each cluster, epsilon(float) = preferred SSE
    stop(int) = stop at this iteration if loop executes too long
    return: tuple of clusters' means(list of np array), clusters' covariance matrices(list of np array), and
        clusters' prior probabilities(list of np array), and SSE(integer)
    '''
    iteration = 0
    K = len(clusters_mean_list)
    curr_prev_SSE = float('inf')
    while curr_prev_SSE > epsilon and iteration < stop:
        iteration += 1
        clusters_weights = [[] for i in range(K)]  # 2d list(outer index: cluster id, inner elem: weight of xj)
        # expectation step
        for j in range(D.shape[0]):  # going through all data points
            xj = D[j]
            denom = 0
            for i in range(2*K):  # going through all clusters
                if i < K:
                    # totaling up the denominator of posterior probability
                    # summing up all f(xj | mu_i, sigma_i)*P(Ci) for all clusters i=1:k
                    denom += prob_xj_given_theta(xj, clusters_mean_list[i], clusters_covmat_list[i])*prior_prob_list[i]
                else:
                    # computing posterior probability
                    ind = i % K
                    # invoking f(x | mu, sigma) function
                    numerator1 = prob_xj_given_theta(xj, clusters_mean_list[ind], clusters_covmat_list[ind])
                    clusters_weights[ind].append(numerator1 * prior_prob_list[ind] / denom)
                #print(f"{i}'s denom: {denom}'")  # debugging purpose
                #print(f"{i}'s weight:\n{clusters_weights}")  # debugging purpose
        print(f"weights:\n{clusters_weights}")  # debugging purpose
        # maximization step
        temp_cluster_mean = []
        for i in range(K):  # iterate through each cluster
            # update cluster mean for each cluster
            wi_vector = np.array(clusters_weights[i])
            denom = np.inner(wi_vector,np.array([1 for x in range(D.shape[0])]))
            updated_mu_i = D.T@wi_vector/denom
            temp_cluster_mean.append(updated_mu_i)
            # update cluster covariance matrix for each cluster
            numerator = np.zeros((D.shape[1], D.shape[1]))
            for j in range(D.shape[0]):
                xj_centered_mu_i = D[j] - temp_cluster_mean[i]  # use the updated mean to compute centered xj
                xj_centered_mu_i.shape = (D.shape[1], 1)
                numerator += wi_vector[j]*(xj_centered_mu_i@xj_centered_mu_i.T)
            updated_sigma_i = numerator/denom
            clusters_covmat_list[i] = updated_sigma_i  # inplace update by replacing previous iteration parameter
            # update prior prob for each cluster
            updated_prior_prob = denom / D.shape[0]  # for updating prior probability, we use <wi,1> in the numerator
            prior_prob_list[i] = updated_prior_prob  # inplace update by replacing previous iteration parameter
        print(f"iteration {iteration}'s temp_cluster_mean:\n{temp_cluster_mean}")
        print(f"iteration {iteration}'s clusters_covmat_list:\n{clusters_covmat_list}")
        print(f"iteration {iteration}'s prior_prob_list:\n{prior_prob_list}")
        # print(f"iteration {iteration}'s Sum of prior prob: {sum(prior_prob_list)}")  # debugging purpose
        # I'm using the Frobenius norm because the Frobenius norm is the square root of sum of abs(entry)^2
        # but entry^2 = (-entry)^2 or (entry)^2. Thus, this would give the same result as SSE induced by l-2 norm
        curr_prev_SSE = np.linalg.norm(np.array(temp_cluster_mean) - np.array(clusters_mean_list), ord='fro')**2
        clusters_mean_list = temp_cluster_mean
        print(f"iteration {iteration}'s SSE between current and previous iteration: {curr_prev_SSE}")
    return clusters_mean_list, clusters_covmat_list, prior_prob_list, curr_prev_SSE
```

```python
#Initialization given in the problem
k = 3
mu_list = [np.array([0.5, 4.2]), np.array([2.4, 1.6]), np.array([3, 3.2])]
covmat_list = []
for i in range(3):
    covmat_list.append(np.identity(D.shape[1]))
prior_probabilities = [1/3, 1/3, 1/3]
```

```python
eps = 0.001
result = expectation_maximization(D, mu_list.copy(), covmat_list.copy(), prior_probabilities.copy(), eps)
#result = expectation_maximization(D, mu_list.copy(), covmat_list.copy(), prior_probabilities.copy(), eps, stop=1)
clusters_mu, clusters_covm, clusters_prior_pr, final_SSE = result
```

Solution to Problem#2(b)
As we can see from the scatter plot in the top left corner([0][1]), The initial means of clutsers are far away from most of the observed data points(Especially Cluster 1 and Cluster 2 means are). This may cause the distance from the cluster mean to each one of data points in cluster to be big(i.e., big mahalanobis distance). Moreover, only two data points are assigned to cluster two because cluster 2's mean c2 is almost at the edge of the plot and its bigger mahalanobis distance contributed to lower Pr(xj | Cluster 2) as a data point is assigned to cluster i based on argmax(of i) f(xj | mu_i, sigma_i). This is confirmed in the cluster labled scatter plot in the bottom left corner.
After first iteration, according to the plot in the top right corner, now first iterated means of clusters are closer to the data points(they are on the edge of the plot as initial means did). Hence, this will reduce the distance between cluster mean and a data point in the data set and result in smaller mahalanobis distance compared to the initial means. And the cluster labled scatter plot in the bottom right corner confirms this fact that the distance between a cluster mean and points assigned to its cluster is smaller than the initial means.
